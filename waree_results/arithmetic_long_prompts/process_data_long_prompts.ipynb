```python
import os
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Get the current directory
current_dir = os.getcwd()

# Initialize an empty dictionary to store the DataFrames
data = {}

# Loop through all the files in the directory
for file in os.listdir(current_dir):
    # Check if the file is a CSV file
    if file.endswith('.csv'):
        # Parse the model name and dataset name from the filename
        model_name, train_dataset, test_dataset, method = file[:-4].split('-')
        
        # Read the CSV file into a pandas DataFrame
        df = pd.read_csv(file)
        
        # Store the DataFrame in the dictionary with keys for the model name and dataset name
        # if model_name not in data:
        #     data[model_name] = {}
        # data[model_name][dataset_name] = df

        data[(model_name, train_dataset, test_dataset, method)] = pd.read_csv(file)
```

## imdb

```python
model = 'llama13b'
train_dataset = 'imdb0'

fig, ax = plt.subplots()
ax.set_title(fr"{model} trained on correctly spelt imdb, evaluated misspelled imdb")
ax.set_xlabel("layer")
ax.set_ylabel("accuracy")
ax.grid(True)

# Set line colors and styles
line_colors = ['black', 'blue', 'red', 'green', 'orange','purple', 'teal', 'pink', 'sage', 'grey']
ccs_line_style = '--'
lr_line_style = ':'
zs_line_style = '-.'

for i, test_dataset in enumerate(['imdb0', 'imdb1', 'imdb2', 'imdb3', 'imdb5', 'imdb10','imdb50']):
    
    experiment = data[(model, train_dataset, test_dataset, 'vinc')]
    layers = experiment['layer']
    acc = experiment['acc']
    # Set line color for current dataset
    line_color = line_colors[i]
    ax.plot(layers, acc, label=f'VINC tested on {test_dataset}', color=line_color)

    experiment = data[(model, train_dataset, test_dataset, 'ccs')]
    layers = experiment['layer']
    acc = experiment['acc']
    ax.plot(layers, acc, label=f'CCS tested on {test_dataset}',
             color=line_color, linestyle=ccs_line_style)

    if 'lm_acc' in experiment.columns:
        lm_acc = experiment['lm_acc']
        # Set line color and style for zero-shot dataset
        ax.plot(layers, lm_acc, label='zero-shot language model output',
                color=line_color, linestyle=zs_line_style)

    if 'lr_acc' in experiment.columns:
        lr_acc = experiment['lr_acc']
        # Set line color and style for logistic regression dataset
        ax.plot(layers, lr_acc, label='logistic regression',
                color=line_color, linestyle=lr_line_style)
                
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
model = 'llama13b'
train_dataset = 'imdb0'

fig, ax = plt.subplots()
ax.set_title(fr"{model} trained on correctly spelt imdb, evaluated misspelled imdb")
ax.set_xlabel("1 mispelling per x characters")
ax.set_ylabel("accuracy")
ax.grid(True)

x = np.array([1,2,3,5,10,20,30,40,50,60])

for i,layer in enumerate([10,20]):
    line_color = line_colors[i]

    acc_vinc = []
    acc_ccs = []
    acc_lr = []
    acc_lm = []

    for test_dataset in ['imdb1', 'imdb2', 'imdb3', 'imdb5', 'imdb10','imdb20','imdb30', 'imdb40', 'imdb50','imdb0']:
        
        experiment = data[(model, train_dataset, test_dataset, 'vinc')]
        acc_vinc.append(experiment.loc[layer, 'acc'])

        experiment = data[(model, train_dataset, test_dataset, 'ccs')]
        acc_ccs.append(experiment.loc[layer, 'acc'])

        acc_lr.append(experiment.loc[layer, 'lr_acc'])

        #if 'lm_acc' in experiment.columns:
            #acc_lm.append(experiment.loc[layer, 'lm_acc'])
            
    ax.plot(x, acc_vinc, label=f'VINC at layer {layer}', color=line_color)
    ax.plot(x, acc_ccs, label=f'CCS at layer {layer}', color=line_color, linestyle=ccs_line_style)
    ax.plot(x, acc_lr, label=f'logistic regression at layer {layer}', color=line_color, linestyle=lr_line_style)
                
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
